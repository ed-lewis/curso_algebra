```python
import numpy as np
```

```python
matriz = [[2, 3, 4],
          [1, 2, 2],
          [5, 3, 3]]
```

```python
matriz1 = np.array(matriz)
matriz1
```

```python
matriz2 = np.array([[2,3,4], [3,3,2], [4,1,6]])
matriz2
```

```python
mult = np.dot(matriz1,matriz2)
print(mult)
```

```python
inversa = np.linalg.inv(matriz1)
print(inversa)
```

```python
inv = np.dot(matriz1, inversa)
print(inv)
```

```python
determinante = np.linalg.det(matriz2)
print(determinante)
```
