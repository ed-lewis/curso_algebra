```python
import numpy as np
import matplotlib.pyplot as plt

v = np.array([1, 2, 3])
s = np.array([3, 2, 1])
print(f"Versão numpy: {np.__version__}")
print(f"Meu vetor: {v}")
```

```python
def produto_interno(v,s):
    prod_interno = 0
    for i in range(len(v)):
        prod_interno = prod_interno + v[i] * s[i]
        i += i + i
    
    return print(f"Produto interno: {prod_interno}")
```

```python
produto_interno(v,s)
```

```python
def produto_escalar(x,y):
    resultado = x * y
    return print(f"Produto escalar: {resultado}")
```

```python
produto_escalar(v,3)
```

# Distância entre vetores

image.png

```python
def distancia (vetor1, vetor2):
    acum = 0
    for i in range (len(v)):
        acum += ( acum + ((vetor1[i]-vetor2[i])**2) )
        print(f'Passo {i}')
    distancia = acum**0.5
    print("Distância entre os vetores: {:.2f}".format(distancia))
```

```python
distancia(v,s)
```

```python
m = np.array([4,5])
```

```python
p = np.array([-2,5])
```

```python
soma = []
for i in range (len(m)):
    soma.append(m[i] + p[i])
    i = i + 1
print(soma)
```

```python
V = np.array([m, p, soma])
origin = np.array([[0,0,0], [0,0,0]])

plt.quiver(*origin, V[:,0], V[:,1], color=['r','b','y'], scale=35)
plt.show();
```
